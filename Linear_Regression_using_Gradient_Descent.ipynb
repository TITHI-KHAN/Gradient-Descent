```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
df=pd.read_csv('data.csv')
df.head()
```

```python
plt.scatter(df.x ,df.y)
```

```python
x = df['x']
y = df['y']
```

**Linear Regression using Gradient Descent**

```python
# 1st step

m = 0 #gradient
c = 0 # y intercept
l = 0.0001 #learning rate
n = int(len(x)) #number of datapoints
iterations = 1000

# 2nd step
for i in range(iterations):
    pred = (m*x) + c
    der_m = (-2/n) *sum(x*(y-pred)) #derivative of m -> der_m #Formula of Step 2

# 3rd step
    der_c = (-2/n) *sum((y-pred))  #derivative of c -> der_c #Formula of Step 3

# 4th step
    m = m-(l*der_m)
    c = c-(l*der_c)

print(m,c) #print m & c outside the loop #getting good weight
```

**Least Squared** (To check whether the prediction from before is right or not)

**R Squared Value / Model Accuracy**

```python
#Way no: 01

reg.score(df[['x']],y) #first predicts, then compares performance
```

```python
#Way no: 02

y_pred = reg.predict(df[['x']]) #Predicted y

from sklearn.metrics import r2_score
Score = r2_score(y, y_pred) #compare performance
```
